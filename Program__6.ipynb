###BAR GRAPH

```python
name = ["A","B","C","D","E","F","S","Z"]
```

```python
salary = np.random.randint(20000,50000,8)
```

```python
salary
```

```python
color_name = ["r","g","b","#0ef0ec","#3b4461","#e87d13","#d124c3","#db0b38"]
```

```python
plt.bar(name,salary, color =color_name, width = 0.8)
plt.xlabel("NAME")
plt.ylabel("SALARY")
plt.title("NAME VS SALARY")
plt.show()
```
